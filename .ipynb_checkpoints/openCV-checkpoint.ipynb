```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import clear_output
```

```python
img = cv2.imread('images/cousines.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# cv2.imshow('image', img)
plt.imshow(img, cmap='gray', interpolation='bicubic')
```

```python
vid = cv2.VideoCapture(0)
try:
    while(True):
        ret, frame = vid.read()
        if not ret:
            vid.release()
            print( "Released Video Resource")
            break
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        plt.imshow(frame, cmap='gray')
        plt.show()
        clear_output(wait=True)
except KeyboardInterrupt:
    vid.release()
    print("Released Video Resource")
```

## Image Arithmetic

```python
img1 = cv2.imread('images/cousines.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('images/marriage.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.imread('images/f#.png')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
```

```python
#add = img1 + img2
add = cv2.add(img1, img2)
plt.imshow(add)
```

```python
weighted_img = cv2.addWeighted(img1, 0.3, img2, 0.7, 0)
plt.imshow(weighted_img)
```
